```python
# Dependencies and Setup
import pandas as pd

import os

# File to Load (Remember to Change These)
file_to_load =  os.path.join ("Resources" , "purchase_data.csv")

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

#display end of data frame information
purchase_data.tail()
```

## Player Count

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#count of unique items
unique_df = purchase_data["Item ID"].nunique()
#unique_df

#average price of purchase
avg_price_df = purchase_data["Price"].mean()
#avg_price_df

#average age of players
avg_age_df = purchase_data["Age"].mean()
#avg_age_df

#total amount of purchases
purchase_amt_df = purchase_data["Item ID"].count()
#purchase_amt_df

#total revenue
revenue_df = purchase_data["Price"].sum()
#revenue_df

#create data frame
purchase_analysis = pd.DataFrame({'Number of Unique Items': [unique_df],
                                 'Average Price': [avg_price_df],
                                 'Average Age': [avg_age_df],
                                 'Number Purchases':[purchase_amt_df],
                                 'Revenue': [revenue_df]})
#format data frame info
purchase_analysis['Average Age'] = purchase_analysis['Average Age'].map('{:,.0f}'.format)
purchase_analysis['Average Price'] = purchase_analysis['Average Price'].map('${:.2f}'.format)
purchase_analysis['Revenue'] = purchase_analysis['Revenue'].map('${:.2f}'.format)

#display data frame
purchase_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#create bins
#bins = [0,9,14,19,24,29,34,39,100]

#create bins second choice
bins = [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,100]
age_groups = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

#reviewing 1st data frame and getting gender, player ID, and age
purchase_data2 = purchase_data.loc[:,['Gender', 'SN', 'Age']]
#purchase_data2
remove_duplicates = purchase_data2.drop_duplicates()

#when cutting sets index by data cutting the data by age information
remove_duplicates['Age Group']= pd.cut(remove_duplicates['Age'], bins,labels=age_groups)
#remove_duplicates["Age Group"]

#count the unique players in each age bracket
age_purchase_count = remove_duplicates['Age Group'].value_counts()
#age_purchase_count.head(20)

#calculate percent of player in each age range
percent_players= age_purchase_count/players_df
#percent_players

#create data frame
age_demo = pd.DataFrame({'Total Count': age_purchase_count,
                        'Percentage of Players': percent_players})
age_demo                                  

#format data format information
age_demo['Percentage of Players'] = age_demo['Percentage of Players'].map('{:,.2%}'.format)
age_demo

#reset index by age
age_demo = age_demo.sort_index()
age_demo
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#create bins
bins = [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,100]
age_groups = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

#obtain source information from main data frame
purchase_data3 = purchase_data.loc[:,['Purchase ID', 'SN', 'Age', 'Price']]
#purchase_data3

#when cutting sets index by age information
purchase_data3["Age Group"] = pd.cut(purchase_data3['Age'], bins,labels=age_groups)
#purchase_data3

#counting the players by age group data cut above
age_group_count = purchase_data3.groupby(['Age Group'])['Purchase ID'].count()
age_group_count

#calculate the average price of the groupby data on price info
avg_price = purchase_data3.groupby(['Age Group'])['Price'].mean()
avg_price

#calculate the total price of the groupby data on price info
tot_price = purchase_data3.groupby(['Age Group'])['Price'].sum()
tot_price

#count unique count purchases per player
unique_playercount = purchase_data3.groupby(['Age Group'])['SN'].nunique()
unique_playercount

#calculate the average total price based on unique players
avg_total = tot_price/unique_playercount
avg_total

#create data frame
purchase_analysis= pd.DataFrame({'Purchase Count': age_group_count,
                                 'Average Purchase Price': avg_price,
                                 'Total Purchase Value': tot_price,
                                 'Avg Total Purchase per Person': avg_total})
purchase_analysis                                  

#format data format information
purchase_analysis['Average Purchase Price'] = purchase_analysis['Average Purchase Price'].map('${:,.2f}'.format)
purchase_analysis['Total Purchase Value'] = purchase_analysis['Total Purchase Value'].map('${:,.2f}'.format)
purchase_analysis['Avg Total Purchase per Person'] = purchase_analysis['Avg Total Purchase per Person'].map('${:,.2f}'.format)
purchase_analysis

#reset index by age
purchase_analysis= purchase_analysis.sort_index()
purchase_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#new data frame with purchase id, player, and price info
purchase_data4 = purchase_data.loc[:,['Purchase ID', 'SN', 'Price']]
#purchase_data4

#grouped data by player ID and counting by purchase ID
top_spender= purchase_data4.groupby(['SN'])['Purchase ID'].count()
#top_spender

#grouped data by player ID and calculating the mean of price 
top_spend_avgprice = purchase_data4.groupby(['SN'])['Price'].mean()
#top_spend_avgprice

#grouped data by player ID and calculating the sum of price 
top_purchase_price = purchase_data4.groupby(['SN'])['Price'].sum()
#top_purchase_price

#create data frame
top_spender_analysis= pd.DataFrame({'Purchase Count':top_spender,
                                 'Average Purchase Price':top_spend_avgprice,
                                 'Total Purchase Value': top_purchase_price})

#top_spender_analysis
top_spender_analysis = top_spender_analysis.sort_values(by=['Purchase Count'], ascending=False)  
#top_spender_analysis

#format data format information
top_spender_analysis['Average Purchase Price'] = top_spender_analysis['Average Purchase Price'].map('${:,.2f}'.format)
top_spender_analysis['Total Purchase Value'] = top_spender_analysis['Total Purchase Value'].map('${:,.2f}'.format)
top_spender_analysis

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns

* Create a summary data frame to hold the results


* Sort the purchase count column in descending order
