```python
# Dependencies and Setup
import pandas as pd

import os

# File to Load (Remember to Change These)
file_to_load =  os.path.join ("Resources" , "purchase_data.csv")

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

#display end of data frame information
purchase_data.tail()
```

## Player Count

* Display the total number of players


```python
#unique() gives all unique values and nunique gives a count of those unique values
players_df = purchase_data["SN"].nunique()
#players_df

#save to new data frame
player_count = pd.DataFrame({"Total Players": [players_df]})

#display data frame
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#count of unique items
unique_df = purchase_data["Item ID"].nunique()
#unique_df

#average price of purchase
avg_price_df = purchase_data["Price"].mean()
#avg_price_df

#average age of players
avg_age_df = purchase_data["Age"].mean()
#avg_age_df

#total amount of purchases
purchase_amt_df = purchase_data["Item ID"].count()
#purchase_amt_df

#total revenue
revenue_df = purchase_data["Price"].sum()
#revenue_df

#create data frame
purchase_analysis = pd.DataFrame({'Number of Unique Items': [unique_df],
                                 'Average Price': [avg_price_df],
                                 'Average Age': [avg_age_df],
                                 'Number Purchases':[purchase_amt_df],
                                 'Revenue': [revenue_df]})
#format data frame info
purchase_analysis['Average Age'] = purchase_analysis['Average Age'].map('{:,.0f}'.format)
purchase_analysis['Average Price'] = purchase_analysis['Average Price'].map('${:.2f}'.format)
purchase_analysis['Revenue'] = purchase_analysis['Revenue'].map('${:.2f}'.format)

#display data frame
purchase_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#total number of players 576
players_df = purchase_data["SN"].nunique()
#players_df

#for each gender in gender column get unique value then once determine all can do percentage of each.
male_df = purchase_data[purchase_data["Gender"]== "Male"]
#male_df

#count amount of males 484
count_male_df = male_df["SN"].nunique()
#count_male_df

#calculate the percentage
male_percentage = count_male_df/players_df
#male_percentage

#total amount of females 81
female_df = purchase_data[purchase_data["Gender"]== "Female"]
#female_df

#count amount of females in team list
count_female_df = female_df["SN"].nunique()
#count_female_df

#calculate the percentage
female_percentage = count_female_df/players_df
#female_percentage

#total amount of other 11
othergender_df = purchase_data[purchase_data["Gender"]== "Other / Non-Disclosed"]
#othergender_df.head()

#count amount of females in team list
count_othergender_df = othergender_df["SN"].nunique()
#count_othergender_df

#calculate the percentage
other_percentage = count_othergender_df/players_df
#other_percentage

#new data frame with calculated information above, index which is gender, columns list [gender, count, percent], dtype=None, copy=False
gender_demo = pd.DataFrame({'Gender': ['Male', 'Female', 'Other / Non-Disclosed'],
                                 'Count': [count_male_df, count_female_df, count_othergender_df],
                                 'Percent': [male_percentage, female_percentage, other_percentage]})

#format data
gender_demo['Percent'] = gender_demo['Percent'].map('{:,.2%}'.format)

#adjust index to gender 
gender_demo = gender_demo.set_index('Gender')

#display dataframe
gender_demo

#possible alternative
#gender_demo = pd.DataFrame(data={[['Male': count_male_df, male_percentage],['Female':count_female_df, female_percentage],['Other': count_othergender_df, other_percentage]]} , index=purchase_data["Gender"], columns=['Gender','Count', 'Percent'], dtype=None, copy=False)
#gender_demo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#male purchase count
male_purchase_count = male_df["Purchase ID"].count()
#male_purchase_count

#male average purchase price
male_avg_price = male_df["Price"].mean()
#male_avg_price

#male total purchase value
male_tot_price = male_df["Price"].sum()
#male_tot_price

#count unique males to count purchases per player
unique_mplayercount = male_df["SN"].nunique()
#unique_mplayercount

#male avg total purchase per person
maleavg_total = male_tot_price/unique_mplayercount 
#maleavg_total


#female purchase count
female_purchase_count = female_df["Purchase ID"].count()
#female_purchase_count
#female_purchase = purchase_data.loc[purchase_data['Gender'] == 'Female', ['Purchase ID', 'Price']]

#female average purchase price
female_avg_price = female_df["Price"].mean()
#female_avg_price

#female total purchase value
female_tot_price = female_df["Price"].sum()
#female_tot_price

#count unique females to count purchases per player
unique_fplayercount = female_df["SN"].nunique()
#unique_fplayercount

#female average total purchase per person
femaleavg_total = female_tot_price/unique_fplayercount 
#femaleavg_total


#other purchase count
other_purchase_count = othergender_df["Purchase ID"].count()
#other_purchase_count

#other average purchase price
other_avg_price = othergender_df["Price"].mean()
#other_avg_price

#other total purchase value
other_tot_price = othergender_df["Price"].sum()
#other_tot_price

#count unique other to count purchases per player
unique_oplayercount = othergender_df["SN"].nunique()
#unique_oplayercount

#other average purchase total per person
otheravg_total = other_tot_price/unique_oplayercount 
#otheravg_total

#create new data frame
purchasing_analysis = pd.DataFrame({'Gender': ['Male', 'Female', 'Other / Non-Disclosed'],
                                    'Purchase Count': [male_purchase_count, female_purchase_count, other_purchase_count],
                                  'Average Purchase Price': [male_avg_price, female_avg_price, other_avg_price],
                                  'Total Purchase Value': [male_tot_price, female_tot_price, other_tot_price],
                                    'Average Total Purchase Per Person': [maleavg_total, femaleavg_total, otheravg_total]})

#format data format information
purchasing_analysis['Average Purchase Price'] = purchasing_analysis['Average Purchase Price'].map('${:,.2f}'.format)
purchasing_analysis['Total Purchase Value'] = purchasing_analysis['Total Purchase Value'].map('${:,.2f}'.format)
purchasing_analysis['Average Total Purchase Per Person'] = purchasing_analysis['Average Total Purchase Per Person'].map('${:,.2f}'.format)

#adjust index to gender
purchasing_analysis = purchasing_analysis.set_index('Gender')

#display new data frame
purchasing_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#create bins
bins = [0,9,14,19,24,29,34,39,100]

age_groups = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

purchase_data['Age Group']= pd.cut(purchase_data['Age'].unique(), bins,labels=age_groups)
#purchase_data["Age Group"].nunique()

ages_df = purchase_data['Age Group'].groupby(['Age Group'])
#ages_df.count()

age_purchase_count = ages_df['SN'].count()
age_purchase_count
```
